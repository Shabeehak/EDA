```python
new_df1_target0=base_data.loc[base_data["NoShow"]==0]
new_df1_target1=base_data.loc[base_data["NoShow"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Hypertension',title='Distribution of Gender for NoShow Customers',hue='Gender')
```

### Insights from EDA on Medical Appointments Data  

The exploratory data analysis (EDA) of medical appointments provided several key insights regarding patient behavior, attendance rates, and influencing factors:  

1. **Gender-Based Trends:**  
   - Female patients have taken more appointments compared to male patients, indicating a higher participation rate in healthcare visits among women.  

2. **No-Show vs. Show Rate Across Age Groups:**  
   - The ratio of no-show and show rates remains almost equal across different age groups, except for infants (Age 0 and Age 1), where the show rate is significantly higher at around 80%.  

3. **Neighborhood-Wise Show Rate:**  
   - Across all neighborhoods, the average show rate remains consistent at approximately 80%, suggesting that location does not have a significant impact on attendance.  

4. **Impact of Scholarship on Attendance:**  
   - Among 99,666 patients without a scholarship, around 80% attended their appointments.  
   - Among the 21,801 patients with a scholarship, attendance was slightly lower at around 75%, indicating that financial assistance does not necessarily lead to better appointment adherence.  

5. **Effect of Hypertension on Show Rates:**  
   - Of the 88,726 patients without hypertension, 78% attended their appointments.  
   - Among the 21,801 patients with hypertension, attendance was higher at 85%, suggesting that patients with hypertension may be more likely to prioritize their health and medical visits.  

6. **Effect of Diabetes on Show Rates:**  
   - 80% of the 102,584 patients without diabetes showed up for their appointments.  
   - Among the 7,943 diabetic patients, the attendance rate was slightly higher at 83%, indicating that those with chronic conditions like diabetes are more committed to medical checkups.  

7. **Impact of SMS Reminders:**  
   - Interestingly, SMS reminders did not seem to improve attendance rates.  
   - Among 75,045 patients who did not receive an SMS, 84% showed up for their appointment.  
   - Among the 35,482 patients who received an SMS reminder, only 72% attended their appointment, indicating that SMS notifications might not be effective in encouraging attendance.  

8. **Appointment Distribution by Day of the Week:**  
   - No appointments were scheduled on Sundays.  
   - Saturdays had significantly fewer appointments compared to weekdays, suggesting that most patients prefer weekday appointments or that weekend scheduling is limited.  

These insights can help improve scheduling strategies, patient outreach methods, and targeted interventions to enhance appointment attendance rates in the future.
