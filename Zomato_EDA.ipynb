## Assignment 
Find the top 10 cuisines

```python
# Get the top 10 cuisines
top_10_cuisines = final_df['Cuisines'].value_counts(sort=True, ascending=False)[:10]
top_10_cuisines
```

```python
# Plot the bar chart
plt.figure(figsize=(12,6))
top_10_cuisines.plot(kind='bar', color='skyblue')

# Formatting
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Cuisines', fontsize=12)
plt.title('Top 10 Cuisines', fontsize=14)
plt.show()
```

```python
# Get the top 10 cuisines with 'Excellent' ratings
top_10_cuisines_excellent = final_df[final_df['Rating text'] == 'Excellent']['Cuisines'].value_counts().nlargest(10)
top_10_cuisines_excellent
```

```python
# Plot the bar chart
plt.figure(figsize=(12,6))
top_10_cuisines_excellent.plot(kind='bar', color='green')

# Formatting
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Cuisines', fontsize=12)
plt.title('Top 10 Cuisines Based on "Excellent" Rating', fontsize=14)
plt.show()
```

# **Insights and Recommendations from Zomato Dataset - Exploratory Data Analysis**  

## **Insights from the Zomato Data Analysis**  

1. **Rating Distribution:**  
   - A significant portion of restaurants have not been rated, indicating that either customers are not leaving reviews or certain restaurants are not receiving enough visibility.  
   - The majority of ratings fall between **2.5 to 3.4**, suggesting that most restaurants provide an average dining experience rather than exceptional service.  

2. **Geographical Trends:**  
   - **India dominates the Zomato dataset**, followed by the **USA and the United Kingdom**, indicating Zomatoâ€™s strong presence in India.  
   - A large number of **0 ratings come from Indian customers**, which could be due to various reasons like a lack of user engagement in leaving feedback or restaurants not encouraging ratings.  

3. **Online Delivery Availability:**  
   - Online delivery services are mainly **available in India and UAE**, suggesting that Zomatoâ€™s delivery operations are more focused on these regions, while other countries primarily use it for restaurant discovery.  

4. **Top Cities for Zomato Restaurants:**  
   - The top five cities with the highest number of restaurants listed on Zomato are:  
     1. **New Delhi**  
     2. **Gurgaon**  
     3. **Noida**  
     4. **Faridabad**  
     5. **Ghaziabad**  
   - This indicates Zomato's stronghold in **North India**, particularly in the **Delhi NCR region**.

## **Recommendations for Zomato Based on Insights**  

1. **Encourage Customer Ratings & Reviews:**  
   - Since a high percentage of restaurants remain unrated, Zomato should introduce **incentives** for customers to leave ratings and reviews, such as discounts or loyalty points.  
   - Send **follow-up notifications** after an order or dining experience to remind users to rate and review.  

2. **Improve Customer Engagement in India:**  
   - Since **many Indian customers do not leave ratings**, Zomato can educate users on the importance of reviews by showcasing **restaurant recommendations based on user feedback**.  
   - Restaurants in India should be encouraged to **ask for reviews** from customers, especially those who frequently order online.  

3. **Expand Online Delivery Services:**  
   - Given that **online deliveries are only available in India and UAE**, Zomato should **expand its delivery services to other countries** where it has a strong presence, such as the USA and UK.  
   - Conduct **market research** to identify high-demand locations where online food delivery could be a profitable expansion.  

4. **Improve Restaurant Quality for Better Ratings:**  
   - Since the **majority of ratings fall between 2.5 and 3.4**, Zomato can collaborate with restaurant owners to **enhance service quality** by providing insights on customer preferences, delivery efficiency, and menu improvements.  
   - Introduce a **restaurant performance dashboard** for owners to track feedback and improve their services accordingly.  

5. **Leverage the Popularity of High-Rated Cuisines:**  
   - As **CafÃ©, North Indian, Italian, Ice Cream, and Seafood** are the most highly-rated cuisines, Zomato should **highlight and promote** restaurants specializing in these cuisines.  
   - Introduce **exclusive deals and promotions** for top-rated cuisines to attract more customers.  

6. **City-Specific Promotions:**  
   - Since most transactions occur in **New Delhi, Gurgaon, Noida, Faridabad, and Ghaziabad**, Zomato can launch **targeted marketing campaigns** and **exclusive food festivals** in these cities to further strengthen its dominance.  

By implementing these strategies, Zomato can enhance user engagement, improve restaurant quality, and expand its online delivery operations for long-term growth.
